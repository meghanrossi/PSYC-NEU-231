```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Add a percentage tip column

```python
tips = sns.load_dataset('tips')
tips.head()
tips['tip_percentage'] = tips['tip']/tips['total_bill']
tips.head()
```

## Visualize the relationship between tip percentage, with day of week and time 

```python
sns.catplot(x = 'day', y = 'tip_percentage', data = tips, kind = 'strip', hue = 'time')
```

```python
tipslunch = sns.catplot(x = 'day', y = 'tip_percentage', data= tips[tips.time == 'Lunch'], hue = 'size')
plt.title('Lunch tips')
plt.show()
```

```python
tipsdinner = sns.catplot(x = 'day', y = 'tip_percentage', data = tips[tips.time == 'Dinner'], hue = 'size')
plt.title('Dinner tips')
plt.show()
```
